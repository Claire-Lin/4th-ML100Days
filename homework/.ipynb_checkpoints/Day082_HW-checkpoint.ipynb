```python
import os
import keras
import itertools
# Disable GPU
os.environ["CUDA_VISIBLE_DEVICES"] = ""
```

```python
train, test = keras.datasets.cifar10.load_data()
```

```python
## 資料前處理
def preproc_x(x, flatten=True):
    x = x / 255.
    if flatten:
        x = x.reshape((len(x), -1))
    return x

def preproc_y(y, num_classes=10):
    if y.shape[-1] == 1:
        y = keras.utils.to_categorical(y, num_classes)
    return y    
```

```python
x_train, y_train = train
x_test, y_test = test

# Preproc the inputs
x_train = preproc_x(x_train)
x_test = preproc_x(x_test)

# Preprc the outputs
y_train = preproc_y(y_train)
y_test = preproc_y(y_test)
```

```python
from keras.layers import Dropout
def build_mlp(input_shape, output_units=10, num_neurons=[512, 256, 128, 64], drp_ratio=0.2):
    input_layer = keras.layers.Input(input_shape)
    
    for i, n_units in enumerate(num_neurons):
        if i == 0:
            x = keras.layers.Dense(units=n_units, activation='relu', name='hidden_layer%d'%(i+1))(input_layer)
            x = Dropout(drp_ratio)(x)
        else:
            x = keras.layers.Dense(units=n_units, activation='relu', name='hidden_layer%d'%(i+1))(x)
            x = Dropout(drp_ratio)(x)
    
    output_layer = keras.layers.Dense(units=output_units, activation='softmax', name='output_layer')(x)
    
    model = keras.models.Model(inputs=[input_layer], outputs=[output_layer])
    return model
```

```python
# Hyperparams
LEARNING_RATE = 1e-3
EPOCHS = 50
BATCH_SIZE = 256
Drop_ratio = [0.1, 0.2, 0.4]
```

```python
results = {}
for i in Drop_ratio:
    print('Experiment with drop ratio %.1f'%i)
    keras.backend.clear_session()
    
    model = build_mlp(input_shape=x_train.shape[1:], drp_ratio=i)
    model.summary()
    
    optimizer = keras.optimizers.Adam(lr=LEARNING_RATE)
    model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer=optimizer)
    
    model.fit(x_train, y_train, 
             epochs=EPOCHS,
             batch_size=BATCH_SIZE,
             validation_data=(x_test, y_test),
             shuffle=True)
    
    # Collect results
    exp_name_tag = 'drop_ratio_%.1f'%i
    results[exp_name_tag] = {'train_loss': model.history.history['loss'],
                            'valid_loss': model.history.history['val_loss'],
                            'train_acc': model.history.history['accuracy'],
                            'valid_acc': model.history.history['val_accuracy']}
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
color_bar = ['C0', 'C1', 'C2']

plt.figure(figsize=(8, 6))
for i, cond in enumerate(results.keys()):
    plt.plot(range(len(results[cond]['train_loss'])), results[cond]['train_loss'], '-', color=color_bar[i], label=cond)
    plt.plot(range(len(results[cond]['valid_loss'])), results[cond]['valid_loss'], '--', color=color_bar[i], label=cond)
plt.title('Loss')
plt.ylim(1, 3)
plt.legend()
plt.show()

plt.figure(figsize=(8, 6))
for i, cond in enumerate(results.keys()):
    plt.plot(range(len(results[cond]['train_acc'])), results[cond]['train_acc'], '-', color=color_bar[i], label=cond)
    plt.plot(range(len(results[cond]['valid_acc'])), results[cond]['valid_acc'], '--', color=color_bar[i], label=cond)
plt.title('Accuracy')
plt.ylim(0, 1)
plt.legend()
plt.show()
```
